```python
#Резюме
def resume_show(resume):
    if len(resume) == 0:
        print()
        print("Резюме отсутствует")
    else:
        for i in range(len(resume)):
            print(resume[i][0], end=": ")
            for j in range(len(resume[i][1])):
                if i == 0:
                        print(resume[i][1][j], end=" ")
                elif i == 1 or i == 2 or i == 3 or i == 4:
                        print(resume[i][1][j], end = ", ")
                elif i == 5 or i == 6:
                    if len(resume[i][1]) > 1:
                        if j == 0:
                            print(f"\n{resume[i][1][j]}")
                        elif j == len(resume[i][1]) - 1:
                            print(resume[i][1][j], end = "")
                        else:
                            print(resume[i][1][j], end = ",\n")
                    else:
                        print(resume[i][1][j], end = "")
            print()
resume = [["ФИО", ["Фальковский", "Дмитрий", "Олегович"]], 
         ["Контакты", ["+77078530910", "dmitriy.02@list.ru"]], 
         ["Владение языками", ["Русский", "Английский"]], 
         ["Владение языками программирования", ["C++", "Java", "Python", "Html, CSS", "JavaScript"]], 
         ["Образование", ["Незаконченное высшее"]]]
print()
print("Вывод")
resume_show(resume)
print()
print("+Новый иностранный язык")
resume[2][1].insert(0, "Немецкий")
resume_show(resume)
print()
print("+Опыт работы")
work_xp = [["Опыт работы", ["Стеллажи.kz, 3 месяца, помощник маркетолога", "Terralink, 6 месяцев, диспетчер"]]]
resume.extend(work_xp)
resume_show(resume)
print()
print("+Новый язык программирования")
resume[3][1].append("Django")
resume_show(resume)
print()
print("Создание копии")
resume_copied = resume.copy()
print()
print("Копия:")
resume_show(resume_copied)
print()
print("Удалить резюме")
resume.clear()
resume_show(resume)
```

```python
#1
def get_lastname():
    print()
    print("Фамилия:")
    lastname = input()
    return lastname
def get_name():
    print()
    print("Имя:")
    name = input()
    return name
def get_year():
    print()
    print("Курс (номер):")
    year = int(input())
    return year
def get_course():
    print()
    print("Дисциплина:")
    course = input()
    return course
def create_list(list1):
    info = []
    info.append(get_name())
    info.append(get_lastname())
    info.append(get_year())
    info.append(get_course())
    list1.append(info)
def sort_list(list1, part, type):
    for i in range(len(list1) - 1):
        for j in range(len(list1) - i - 1):
            if list1[j][part] > list1[j + 1][part]:
                temp = list1[j]
                list1[j] = list1[j + 1]
                list1[j + 1] = temp
    return list1
def show_list(list1):
    for i in range(len(list1)):
        print(str(i+1) + ") " + list1[i][0] + " " 
              + list1[i][1] + " — " + str(list1[i][2]) + " курс, " 
              + str(list1[i][3]))
list1 = []
print()
print("Количество студентов:")
quantity = int(input())
for i in range(quantity):
    create_list(list1)
print()
print("Неотсортированный список студентов:")
show_list(list1)
print()
print("Список студентов по имени:")
list1 = sort_list(list1, 0, 0)
show_list(list1)
print()
print("Список студентов по курсу:")
list1 = sort_list(list1, 2, 0)
show_list(list1)
```

```python
#2
def show_grades(grades):
    for i in range(len(grades)):
        for j in range(len(grades[0])):
            if j == 0:
                print()
                print("Список студента " + grades[i][j] + ":")
            else:
                print(grades[i][j][0] + ": " + str(grades[i][j][1]))
def check(grades, name):
    for i in range(len(grades)):
        if name in grades[i]:
            return grades[i]
        else:
            print("Студент отсутсвет")
def show_student(grades, name):
    for i in range(len(grades)):
        if name in grades[i]:
            for j in range(len(grades[i])):
                if j == 0:
                    print()
                    print("Список оценок студента " + grades[i][j] + ":")
                else:
                    print(grades[i][j][0] + ": " + str(grades[i][j][1]))

student1 = ["Дмитрий", ["Функциональное программирование", 3.7], ["РМП", 3.5], ["Физра", 0]]
student2 = ["Маргарита", ["Функциональное программирование", 3.8], ["РМП", 3.9], ["Физра", 3.5]]
student3 = ["Диана", ["Функциональное программирование", 3.6], ["РМП", 3.3], ["Физра", 3.8]]
student4 = ["Дильшат", ["Функциональное программирование", 0], ["РМП", 0], ["Физра", 4]]
grades = [student1, student2, student3, student4]
print()
print("Список оценок:")
show_grades(grades)
print()
print("Введите имя избранного:")
name = input()
chosen = check(grades, name)
print()
print(f"Список студента {name}: {chosen}")
show_student(grades, name)
```

```python
#6
def sorted_list(list):
    sorted_asc = sorted(list)
    sorted_desc = sorted(list, reverse = True)

    if sorted_asc == list or sorted_desc == list:
        return True
    else:
        return False
list1 = []
print()
print("Размер списка: ")
n = int(input())
print()
print("Введите элементы списка:")
for i in range(n):
    list1.append(int(input()))
print()
print(f"Ваш список: {list1}")
print()
print(f"Сортировка - {sorted_list(list1)}")
```
