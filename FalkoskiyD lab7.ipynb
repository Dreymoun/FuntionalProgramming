```python
# создание словаря с информацией о студентах
students = {
    "Дмитрий": {
        "возраст": 21,
        "специальность": "Computer Science",
        "gpa": 3.8,
        "опыт работы": ["3 месяца в Стеллажи.kz", "6 месяцев в Terralink"],
        "умения": ["Python", "JavaScript", "Django"]
    },
    "Маргарита": {
        "возраст": 20,
        "специальность": "Computer Science",
        "gpa": 3.6,
        "опыт работы": ["Пишет фанфики", "и больше ничего :D"],
        "умения": ["Java", "Python", "HTML/CSS"]
    },
    "Боб": {
        "возраст": 20,
        "специальность": "Инженер Механик",
        "gpa": 3.2,
        "опыт работы": ["Опыт работы в компании ТЕСЛА", "Ассистент"],
        "умения": ["Коммуникабельность", "MATLAB", "Машиностроение"]
    }
}

# функция для вывода информации о студентах
def print_students(students_dict):
    for name, info in students_dict.items():
        print(f"\n{name}:")
        print(f"Возраст: {info['возраст']}")
        print(f"Специальность: {info['специальность']}")
        print(f"GPA: {info['gpa']}")
        print("Опыт работы:")
        for exp in info['опыт работы']:
            print(f"- {exp}")
        print("Умения:")
        for skill in info['умения']:
            print(f"- {skill}")

# функция для добавления информации о студенте
def add_student(students_dict, name, age, major, gpa, experience, skills):
    students_dict[name] = {
        "возраст": age,
        "специальность": major,
        "gpa": gpa,
        "опыт работы": experience,
        "умения": skills
    }
    print(f"{name} успешно добавлен!")

# функция для удаления информации о студенте
def delete_student(students_dict, name):
    del students_dict[name]
    print(f"{name} успешно удален!")

# функция для обновления информации о студенте
def update_student(students_dict, name, key, value):
    students_dict[name][key] = value
    print(f"{name}'s {key} успешно обновлен!")

# функция для поиска информации о студенте
def search_student(students_dict, name):
    print("Поиск...")
    if name in students_dict:
        print_students({name: students_dict[name]})
        print("")
        print("Найден!")
    else:
        print(f"{name} не найден в базе.")

# вывод информации о всех студентах
print_students(students)
print("----------------------")
# добавление нового студента
add_student(students, "Алиса", 19, "Биолог", 3.9, ["Научные сотрудник Гарварда"], ["Лабораторная техника", "Анализ данных"])
print("----------------------")
# удаление информации о студенте
delete_student(students, "Боб")
print("----------------------")
# обновление информации о студенте
update_student(students, "Маргарита", "gpa", 3.7)
print("----------------------")
# поиск информации о студенте
search_student(students, "Дмитрий")
print("----------------------")
print_students(students)
```

```python
print("Введите количесвто рек ")
n = int(input())  # число элементов в словаре

# создание словаря со странами и реками
print(f"Введите {n} стран и рек, которые в ней протекают")
countries_rivers = {}
for i in range(n):
    country, *rivers = input().split()
    countries_rivers[country] = rivers

# список рек для проверки наличия в словаре
print("Какую реку нужно найти?")
rivers_to_check = input().split()

# вывод стран, в которых протекают заданные реки
for river in rivers_to_check:
    found = False
    for country, rivers in countries_rivers.items():
        if river in rivers:
            found = True
            print(f"{river} протекает в {country}")
    if not found:
        print(f"{river} не найдена в записях.")

# добавление новой пары страна-река в словарь
country, river = input().split()
if country in countries_rivers:
    countries_rivers[country].append(river)
    print(f"{river} добавлена {country} успешно!")
else:
    countries_rivers[country] = [river]
    print(f"{country} добавлена в записи с {river}.")
```

```python
commentators = {}  # словарь для хранения количества комментариев у каждого комментатора
print("Пишите комментарии, господа: ")
while True:
    line = input()
    if not line:  # пустая строка означает конец ввода комментариев
        break
    name, comment = line.split(": ")
    if name in commentators:
        commentators[name] += 1
    else:
        commentators[name] = 1

print(f"Количество комментаторов: {len(commentators)}")
```

```python
print("Введите кол-во телефонов в книжке")
n = int(input())
phones = {}
print("Введите номер телефона, затем имя контакта")
for i in range(n):
    phone, name = input().split()
    phones[name] = phone
print("Поиск...")
query = input()
if query in phones:
    print(phones[query])
else:
    print("Нет в телефонной книге")
```

```python
print("Введите количество сотрудников: ")
n = int(input())
vacations = {}
print("Введите сотрудника, день и месяц его отпуска")
for i in range(n):
    surname, day, month = input().split()
    if month in vacations:
        vacations[month].append(surname)
    else:
        vacations[month] = [surname]
print("Введите месяц, по которому требуется найти сотрудника")
requested_month = input()

if requested_month in vacations:
    print(" ".join(vacations[requested_month]))
else:
    print()
```
